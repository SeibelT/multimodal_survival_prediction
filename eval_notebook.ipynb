```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

f_mm_full = "/nodes/bevog/work4/seibel/multimodal_survival_prediction/evaluations/MultiModal_FullTraiined_sweep.csv"
f_mm_nfull  = "/nodes/bevog/work4/seibel/multimodal_survival_prediction/evaluations/MultiModal_notfullytrained_sweep.csv"
f_hist_full = "/nodes/bevog/work4/seibel/multimodal_survival_prediction/evaluations/UnimodalHisto_fulltrained_sweep.csv"

```

```python
df_mm_full = pd.read_csv(f_mm_full)
df_mm_nfull = pd.read_csv(f_mm_nfull)
df_mm_nfull["d_hidden"]=256

df_hist_full = pd.read_csv(f_hist_full)
```

```python
import json

def dimhist_splitter(df):
    keys_list = ["Name","d_hidden","dim_hist_and_feature_path","learningrate", "modality","repeating", "dropout","test/c_index"]
    df_new = df[keys_list].copy()
    df_new["hist_dimension"] = df["dim_hist_and_feature_path"].apply(lambda x : json.loads(x)[0])
    df_new["encoding_name"] = df["dim_hist_and_feature_path"].apply(lambda x : json.loads(x)[1].split("/")[-2])
    df_new = df_new.drop(columns="dim_hist_and_feature_path")
    return df_new
 
```

```python
df_mm_full = dimhist_splitter(df_mm_full)
df_mm_nfull = dimhist_splitter(df_mm_nfull)
df_hist_full = dimhist_splitter(df_hist_full)
```

```python
df_hist_full.groupby(by=["d_hidden","hist_dimension","encoding_name","learningrate", "modality", "dropout"]).mean("test/c_index")
```

```python

df_mm_full.groupby(by=["d_hidden","hist_dimension","encoding_name","learningrate", "modality", "dropout"]).agg({'test/c_index':['mean','std'],'repeating':"max"}).sort_values([('test/c_index', 'mean')],ascending=False).head(10)
```

```python
df_all = pd.concat([df_mm_full , df_mm_nfull , df_hist_full])
df_bestlist = df_all.groupby(by=["d_hidden","hist_dimension","encoding_name","learningrate", "modality", "dropout"]).agg({'test/c_index':['mean','std'],'repeating':"max"}).sort_values([('test/c_index', 'mean')],ascending=False).reset_index()
df_bestlist.encoding_name = df_bestlist.encoding_name.apply(lambda x :x.replace("vit_mae_tiny_I1K","") if x.startswith("vit_mae_tiny_I1K") else x )
df_bestlist.head(60)
```
